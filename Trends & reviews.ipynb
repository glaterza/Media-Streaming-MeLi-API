#### análisis de tendencias

Tengo los top 100 strings de búsqueda en la categoría Media Streaming. Voy a extraer el nombre del producto, para poder hacer el count de apariciones y ver cuáles son los productos más buscados.

```python
trends['keyword'] = trends['keyword'].str.lower()
```

```python
def product_tag(keyword):
    devices = ['chromecast','amazon fire','apple tv','android','xiaomi','roku']
    for device in devices:
            if device in keyword:
                return device
```

```python
trends['product'] = trends['keyword'].apply(product_tag)
```

```python
devices = ['MLA13841393', 'MLA13841392', 'MLA14141341', 'MLA8972016', 'MLA13877782','MLA14141529','MLA14540856',
           'MLA6001887', 'MLA8908816', 'MLA14141340', 'MLA14141529', 'MLA6253551', 'MLA8972016', 'MLA8908816', 
           'MLA13877782', 'MLA6253543', 'MLA6253545', 'MLA6125038']

# Esto lo importé de Tableau.
```

```python
reviews_columns = ['catalog_product_id','rating_average','rating_levels.one_star', 'rating_levels.two_star', 'rating_levels.three_star',
                    'rating_levels.four_star', 'rating_levels.five_star']
```

```python
all_reviews = pd.DataFrame(columns = reviews_columns)
```

```python
url = 'https://api.mercadolibre.com/reviews/item/'
for device in devices:
    response = requests.get(url+device)
    item_reviews = json.loads(response.text)
    reviews_normalized = pd.io.json.json_normalize(item_reviews, errors = 'ignore')
    reviews_normalized['catalog_product_id'] = device
    reviews_normalized = reviews_normalized[['catalog_product_id','rating_average','rating_levels.one_star', 'rating_levels.two_star', 'rating_levels.three_star',
                    'rating_levels.four_star', 'rating_levels.five_star']]
    all_reviews = pd.concat([all_reviews, reviews_normalized])
```
